# 01 EDA

Khám phá dữ liệu raw từ Tiki: volume crawl, phân phối rating, độ dài review và top sản phẩm.


```python
import os
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

engine = create_engine(os.getenv('TIKI_DATA_DB', 'postgresql+psycopg2://airflow:airflow@localhost:5432/tiki_data'))
reviews = pd.read_sql('SELECT rating, product_id, content, purchased, crawled_at FROM raw_reviews', engine)
reviews.head()
```

```python
reviews['content_length'] = reviews['content'].fillna('').str.len()
fig = px.histogram(reviews, x='rating', color='purchased', title='Phân phối rating')
fig.write_html('../reports/figures/rating_distribution.html')
fig
```
