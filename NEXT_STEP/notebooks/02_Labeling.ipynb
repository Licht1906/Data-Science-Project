# 02 Labeling

Kiểm tra weak labels và các heuristic flags.


```python
import os
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

engine = create_engine(os.getenv('TIKI_DATA_DB', 'postgresql+psycopg2://airflow:airflow@localhost:5432/tiki_data'))
processed = pd.read_sql('SELECT is_fake, flag_count, flags, processed_at FROM processed_reviews', engine)
processed.head()
```

```python
fig = px.histogram(processed, x='flag_count', color='is_fake', title='Số lượng heuristic flags')
fig.write_html('../reports/figures/flag_distribution.html')
fig
```
