# 05 NLP Analysis

Phân tích text và interpretability. Nếu có model tree-based, có thể dùng SHAP để giải thích feature importance.


```python
import os
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

engine = create_engine(os.getenv('TIKI_DATA_DB', 'postgresql+psycopg2://airflow:airflow@localhost:5432/tiki_data'))
df = pd.read_sql('SELECT content_clean, is_fake, flag_count FROM processed_reviews', engine)
df['word_count'] = df['content_clean'].fillna('').str.split().str.len()
fig = px.histogram(df, x='word_count', color='is_fake', title='Độ dài review theo nhãn')
fig.write_html('../reports/figures/review_word_count_by_label.html')
fig
```
